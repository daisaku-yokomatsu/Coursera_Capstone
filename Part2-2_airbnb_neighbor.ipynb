### Limit the data to properties that meet the following criteria
- Property_type: Entire apartment or Entire condominium or Entire serviced apartment
- review_scores_rating >= 95
- maximum_nights >= 90

```python
df0= df0[(df0['review_scores_rating'] >= 95) & (df0['maximum_nights'] >= 90)]
```

```python
df0.shape
```

```python
list = ['Entire apartment','Entire condominium','Entire serviced apartment']
df = df0[df0.property_type.isin(list)]
```

```python
df.shape
```

#### The number of apartments that satisfy the conditions is 6036.

```python
st0 = pd.read_csv('DOITT_SUBWAY_STATION_01_13SEPT2010.csv',encoding='UTF-8')
```

```python
st0.head()
```

```python
st0.drop(columns=['URL','OBJECTID','NOTES'],inplace=True)
st = st0.rename(columns={'NAME': 'Station_Name'})
```

```python
  st['lat'] = 0.0
  st['lng'] = 0.0
  st['numuni'] = 0

for j in range(len(st)):
   nam = st['Station_Name'][j] 
   p = st['the_geom'][j]
   x = float(p.split(" ")[1].strip('('))
   y = float(p.split(" ")[2].rstrip(')'))
   st.loc[st.Station_Name == nam,'lat'] = y
   st.loc[st.Station_Name == nam,'lng'] = x
```

```python
st.drop(columns=['the_geom'],inplace=True)
st.head()
```

#### Using the location information of the apartments that meet the conditions and <br>the location information of the subway station, count the number of apartments <br>within 300 meters of the station.

```python
for i in range (len(st)):
  nol = 0
  stp = (st.iloc[i].lat,st.iloc[i].lng)
  for j in range(len(df)):
     bnbp =(df.iloc[j].latitude,df.iloc[j].longitude)
     dis = geodesic(stp,bnbp).m
     #print(i,j,dis)
     if dis <= 300 :
         nol = nol + 1
  print('***',i,nol)
  st.iloc[i,4] = nol
```
